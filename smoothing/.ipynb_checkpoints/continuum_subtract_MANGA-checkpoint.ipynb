<h1>Continuum subtraction - MaNGA cube</h1>

Example of use of slicing, multi-dimension arithmetic, and smoothing.

```python
# All imports used in this script
from __future__ import print_function

import numpy as np
import numpy.ma as ma

import matplotlib.pyplot as plt
import matplotlib.cm as cm

from scipy.interpolate import LSQUnivariateSpline, InterpolatedUnivariateSpline

from astropy.io import fits

%matplotlib inline
plt.rcParams['figure.figsize'] = 8, 10
```

<h3>1 - Get raw data cube</h3>

Read cube from FITS file.

```python
cube = fits.getdata('./data/manga-7443-12703-LOGCUBE.fits')
print(cube.shape)
```

Ensure all NaN-valued pixels are zeroed before proceeding.

```python
cube[np.where(np.isnan(cube))] = 0.
```

Display on image plane (collapse over spectral dimension).

```python
plt.rcParams['figure.figsize'] = 5,5
fig, ax = plt.subplots()
ax.imshow(np.sum(cube,axis=0), vmax=5000, cmap=cm.hot, interpolation='nearest')
```

<h3>2 - Find line emission</h3>

To maximize SNR, build sum from entire cube along wavelength dimension.

```python
summed = np.sum(cube, axis=2)
summed = np.sum(summed, axis=1)

plt.gcf().set_size_inches(16.,4.)
plt.axis([0.,4600.,-100.,1000.])
plt.plot(summed)
```

<h3>3 - Mask out emission regions </h3>

1 - Build mask that rejects wavelength ranges with emission.

2 - Use mask to build a masked data cube.

```python
mask = cube.copy()
mask.fill(False)
mask[ 240: 600,:,:] = True  # these were determined by examining the plot above with a large
mask[ 670: 730,:,:] = True  # zoom factor. Some interaction is unavoidable here.
mask[ 900:1000,:,:] = True  
mask[1400:1600,:,:] = True  
mask[1870:1900,:,:] = True  
mask[2070:2100,:,:] = True  
mask[2250:2310,:,:] = True  
mask[2560:2590,:,:] = True  
mask[2700:2900,:,:] = True  
mask[3450:3500,:,:] = True  
mask[3800:4000,:,:] = True  
mask[4050:4300,:,:] = True  
mask[4310:4563,:,:] = True  

masked_cube = ma.masked_array(cube, mask=mask)
```

Note that we built a mask that is identical for every pixel over the entire spatial extent of the image. But the mask itself is more general and could accomodate any type of spatially-variant masking function. 

Most of the remaining code in this script assumes a spatially-invariant mask.

Note also that the mask includes some absorption lines as well.

<h3>4 - Verify that masked data cube is OK </h3>

Again, build sum for entire *masked* cube. Data in masked wavelength ranges should be absent.

```python
summed_masked = np.sum(masked_cube, axis=2)
summed_masked = np.sum(summed_masked, axis=1)

plt.gcf().set_size_inches(12.,4.)
plt.axis([0.,4600.,-100.,1000.])
plt.plot(summed_masked)
```

<h3>5 - Build 1-D continuum model by interpolation-smoothing summed masked spectrum </h3>

<h4>Scipy interpolation routines work in non-intuitive ways with masked arrays.</h4>

The problem seems to be that masked arrays, from the interpolator's viewpoint, are not really masked. The masked data points seem to be still in the array, not removed from it, and behave as having beign assigned a value of zero. 

Somehow these masked data points interfere with the spline fit, even though we use a set of nodes that explicitly ignores them.

Here is an example usng the only interpolator that allows settable nodes:

```python
# set spline nodes 100 spectral pixels apart
nodes = range(1,4563,100)    

# remove nodes that are inside masked regions
indices_bad = np.where(summed_masked == 0.0)[0].tolist()
good_nodes = [node for node in nodes if node not in indices_bad]
good_nodes = np.array(good_nodes)
print(good_nodes)
```

```python
x = np.asarray(range(len(summed_masked)))

continuum = LSQUnivariateSpline(x, summed_masked, good_nodes)(x)

plt.gcf().set_size_inches(12.,4.)
plt.axis([0.,4600.,-100.,1000.])
plt.plot(summed_masked, 'y')
plt.plot(continuum, 'r')

plt.plot(good_nodes, np.ones(shape=good_nodes.shape), 'bo')
```

This is pretty much the same behavior we get from the other scipy spline interpolators. In those other interpolators we have no control where the nodes are exactly, so the situation is even worse. 

The scipy polynomial interpolators either crash, or exactly reproduce the input data where it is not masked (which is expected) but generate zeros for the interpolated values in the masked regions. So they essentially do not interpolate at all.

Now, the exact same spline fit but with the un-masked data shows more promising results. This suggests that masked data arrays are of no use for this specific purpose.

```python
x = np.asarray(range(len(summed_masked)))

continuum = LSQUnivariateSpline(x, summed, good_nodes, k=3)(x)

constant = np.ones(shape=good_nodes.shape)
plt.gcf().set_size_inches(12.,4.)
plt.axis([0.,4600.,-100.,1000.])
plt.plot(summed_masked, 'y')
plt.plot(continuum, 'r')
plt.plot(good_nodes, constant, 'bo')
```

Better than with the masked array, but somehow the data in the gaps of the nodes array appears to influence the result. It could be as well that node placement is critical. Here, the region between data points 2600 and 2900 is badly influencing the fit, resulting in excess continuum there. 

Running some experiments with nodes defined by hand resulted in similar results, suggesting that it is indeed the data in the nodes gaps, more than node placement, that is influencing the results.

<h4>Scipy interpolation functions do not seem to like weighted fits where the weights are zeros.</h4>

We can try to use weights in the spline interpolation function calls. Additionally, or instead, of using a masked array, we can build a weight array that is one where the data is good, and zero where we want to discard data.

The LSQUnivariateSpline interpolator does not accept weights. The other scipy spline interpolators do accept weights, but the result is always the same. They return an array of NaN's. From the documentation in scipy, we can infer that weights are not supposed to be zero. They can be either 1.0, or a value that relates to the inverse errors in the data.

Here is an example:

```python
indices_bad = np.where(summed_masked == 0.0)
weights = np.ones(len(summed_masked))
weights[indices_bad] = 0.0
```

```python
plt.gcf().set_size_inches(12.,4.)
plt.axis([0.,4600.,-0.1,1.5])
plt.plot(weights, 'g')
```

Now interpolate as above, but adding the weights array to the mix:

```python
continuum = InterpolatedUnivariateSpline(x, summed_masked, w=weights)(x)
```

```python
continuum
```

```python
plt.gcf().set_size_inches(12.,4.)
plt.axis([0.,4600.,-100.,1000.])
plt.plot(summed_masked, 'y')
plt.plot(continuum, 'r')
```

<h4>Doesn't work here either: using weights as above, on a plain, non-masked array, does not solve the problem.</h4>

We can ignore the masked data entirely and use the raw data with weights instead. Maybe masked arrays and weights do not play well together? 

Lets repeat the computations but now using the plain cube as input:

```python
summed = np.sum(cube, axis=2)
summed = np.sum(summed, axis=1)
```

```python
plt.gcf().set_size_inches(12.,4.)
plt.axis([0.,4600.,-100.,1000.])
plt.plot(summed)
```

```python
continuum = InterpolatedUnivariateSpline(x, summed, w=weights)(x)
```

Again, we get an array of NaN's:

```python
continuum
```

```python
plt.gcf().set_size_inches(12.,4.)
plt.axis([0.,4600.,-100.,1000.])
plt.plot(summed, 'y')
plt.plot(continuum, 'r')
```

<h3>6 - Final processing: subtract continuum from cube </h3>

Here, we pick the solution from above that worked the best (non-masked array, nodes at selected data points), and subtract the continuum thus calculated from the original cube.

<b>Compute continuum (as previously done above):</b>

```python
continuum = LSQUnivariateSpline(x, summed, good_nodes)(x)

plt.gcf().set_size_inches(12.,4.)
plt.axis([0.,4600.,-100.,1000.])
plt.plot(summed, 'y')
plt.plot(continuum, 'r')
```

<b>Subtract contiuum from cube</b>

The interpolated continuum spectrum includes the masked regions (that's the whole point of using interpolation in the first place). So we first have to build a masked version of the interpolated continuum, in order to get its flux over the unmasked regions only:

```python
masked_continuum = ma.masked_array(continuum, mask=mask[:,0,0]) # in this example, any spatial pixel should do
continuum_flux = np.sum(masked_continuum)
```

Before subtracting the continuum from each spatial pixel, it has to be re-normalized to the flux in that spatial pixel. Note that we use the masked cube for the computation of the flux in one pixel, so it can be directly related to the flux in the masked interpolated continuum computed above.

```python
continuum_subtracted_cube = np.zeros(shape=cube.shape) 
# these two loops scan the entire spatial extent of the cube. 
for i in range(cube.shape[1]):
    for j in range(cube.shape[2]):
        pixel_flux = np.sum(masked_cube[:,i,j])
        continuum_subtracted_cube[:,i,j] = cube[:,i,j] - (continuum / continuum_flux * pixel_flux)
```

<b>Verify results:</b>

Apart from noise and fitting errors, the final continuum-subtracted image should contain only the flux from emission line wavelength regions (and perhaps negative flux from absorption lines if they were included in the mask). 

```python
print(np.min(continuum_subtracted_cube))
print(np.max(continuum_subtracted_cube))
```

```python
print(np.min(np.sum(continuum_subtracted_cube, axis=0)))
print(np.max(np.sum(continuum_subtracted_cube, axis=0)))
```

```python
plt.rcParams['figure.figsize'] = 5,5
fig, ax = plt.subplots()
ax.imshow(np.sum(continuum_subtracted_cube, axis=0), vmin=-100, vmax=700, cmap=cm.hot, interpolation='nearest')
```

We can verify that globally the cube had its continuum removed by plotting its spectrum over the entire spatial extent of the cube:

```python
summed_cs = np.sum(continuum_subtracted_cube, axis=2)
summed_cs = np.sum(summed_cs, axis=1)

plt.gcf().set_size_inches(16.,4.)
plt.axis([0.,4600.,-500.,600.])
plt.plot(summed_cs)
plt.plot([0.,4600.], [0., 0.], 'r')
```

<h3>7 - Conclusions </h3>

- interpolation with scipy functions is tricky and doesn't work in an intuitive way.
- the only 1-D spline interpolator that allows directly settable nodes is LSQUnivariateSpline.
- use of masked arrays seems to be innefective with the scipy spline interpolation functions.
- figuring out the regions to be masked and the spline node placement is work-intensive.
- once, and if, the interpolation is completed satisfactorily, subtracting the continuum is fairly straightforward.
